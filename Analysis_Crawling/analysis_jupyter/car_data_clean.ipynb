```python
df_2324 = pd.read_excel("20242023.xlsx", skiprows=5)
df_2324.drop(df_2324.columns[[0,2,3,4,5,6,7,8,9]], axis=1, inplace=True)
df_2324.columns = ["브랜드명", "2024등록대수", "2024점유율", "2023등록대수", "2023점유율", "2023대비2024변화율"]

df_2324
```

```python
dataframes = [df_1415, df_1617, df_1819, df_2021, df_2223, df_2324]
merged_df = reduce(lambda left, right: pd.merge(left, right, on="브랜드명"), dataframes)
merged_df = merged_df.rename(columns={"2023등록대수_y": "2023등록대수", "2023점유율_y": "2023점유율"})

merged_df.drop(merged_df.columns[[21,22]], axis=1, inplace=True)
```

```python
merged_df
```

```python
columns_to_drop = [col for col in merged_df.columns if "변화율" in col]
merged_df = merged_df.drop(columns=columns_to_drop)

merged_df.head()
```

```python
merged_df.info()
```

```python
change_df = pd.DataFrame()
change_df["브랜드명"] = merged_df["브랜드명"]

# 변화율 계산 함수
def calculate_change_rate(current, previous):
    return ((current - previous) / previous) * 100

# 등록대수 변화율 계산
for year in range(2014, 2024):  
    current_col = f"{year + 1}등록대수"
    previous_col = f"{year}등록대수"
    if current_col in merged_df.columns and previous_col in merged_df.columns:
        change_col = f"{year}-{year + 1}등록대수변화율"
        change_df[change_col] = calculate_change_rate(
            merged_df[current_col].str.replace(",", "").astype(float),
            merged_df[previous_col].str.replace(",", "").astype(float),
        ).round(1)

# 점유율 변화율 계산
for year in range(2014, 2024):  # 연도 범위 지정
    current_col = f"{year + 1}점유율"
    previous_col = f"{year}점유율"
    if current_col in merged_df.columns and previous_col in merged_df.columns:
        change_col = f"{year}-{year + 1}점유율변화율"
        change_df[change_col] = calculate_change_rate(
            merged_df[current_col].astype(float),
            merged_df[previous_col].astype(float),
        ).round(1)

change_df
```
